## Importing libraries

```python
import pandas as pd
import nltk
import spacy
nltk.download('stopwords')
# To remember which versions I used
print(f'nltk version: {nltk.__version__}')
print(f'spacy version: {spacy.__version__}')
```

## Defining custom function to clean data

```python
# I created this function for another project. Copy-pasting it here
# This function uses code generated by GitHub Copilot

nlp = spacy.load('en_core_web_sm')

def parse_text(input_text):
    '''
    Function to parse text for natural language processing.

    Input: text (string)
    Output: list of words (list)

    Dependencies: nltk, spacy

    If the input is not a string, the function will raise a TypeError.
    If the input is an empty string, the function will return an empty list.
    The function uses stop words from the nltk library for English.
    '''
    # Check that the input is a string
    if type(input_text) != str:
        raise TypeError('The input must be a string.')
    # Check that the input is not an empty string
    if input_text == '':
        return []
    # Convert the input text to lower case
    output_text = input_text.lower()
    # Remove punctuation and tokenize words
    output_text = nltk.tokenize.RegexpTokenizer(r'\w+').tokenize(output_text)
    # Remove numbers
    output_text = [word for word in output_text if not word.isdigit()]
    # Remove stop words
    output_text = [word for word in output_text if word not in nltk.corpus.stopwords.words('english')]
    # Lemmatize words
    output_text = [nlp(word)[0].lemma_ for word in output_text]
    # Return the output text
    return output_text

if __name__ == '__main__':
    # Test function
    # Empty string as input
    assert parse_text('') == []
    # Stop words
    assert parse_text('This is a test') == ['test']
    # Punctuation and tokenization
    assert parse_text('This is a test.') == ['test']
    assert parse_text('This is a test. 123!') == ['test']
    assert parse_text('This is a test. 123!@#') == ['test']
    assert parse_text('This\nis some :) weird\ttext.  DID it work:/?') == ['weird', 'text', 'work']
    # Capitalization
    assert parse_text('This is a Test.') == ['test']
    # Repeated words
    assert parse_text('This is a test. This is only a test.') == ['test', 'test']
    # Digits
    assert parse_text('This is a test. 123') == ['test']
    assert parse_text('This is a test.123') == ['test']
    # assert parse_text('This is a test.123nana') == ['test']
    # Lemmatization
    assert parse_text('Is this working?') == ['work']
    assert parse_text('running') == ['run']
    assert parse_text('jumps') == ['jump']
    assert parse_text('better') == ['well']
    assert parse_text('mice') == ['mouse']
    # Works with emojis
    assert parse_text('hello 😢') == ['hello']
```

## Reading data

```python
df = pd.read_csv('reddit_mindfulness.csv', index_col = 0)
```

## Parsing posts and titles

```python
# Creating new columns for parsed text
df['title_parsed'] = None
df['selftext_parsed'] = None

# Dictionary to keep track of errors while parsing text
errors = {
    'column': [],
    'error': []
}

# Parsing text and populating new columns
for i, row in df.iterrows():
  # Error handling in case my function fails
  try:
    df.at[i, 'title_parsed'] = parse_text(row['title'])
  except Exception as e:
    errors['column'].append('title')
    errors['error'].append(e)
  # Doing it separately for selftext since they could have different errors
  try:
    df.at[i, 'selftext_parsed'] = parse_text(row['selftext'])
  except Exception as e:
    errors['column'].append('selftext')
    errors['error'].append(e)
```
